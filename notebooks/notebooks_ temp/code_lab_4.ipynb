```python
import pandas as pd

functional_df = pd.read_csv("../../data/FIsInGene_070323_with_annotations.txt", delimiter="\t")
```

```python
my_dict = {
    "inhibit": set(),
    "activate": set()
}
def fix_type(el):
    if "inhi" in el or "repr" in el:
        my_dict["inhibit"].add(el)
        return "inhibit"
    elif "activ" in el:
        my_dict["activate"].add(el)
        return "activate"
    else:
        return el

xx = []
for x in functional_df["Annotation"].unique():
    x = x.split("; ")
    if type(x) == list:
        for el in x:
            el = fix_type(el)
            xx.append(el)
```

```python
print(set(xx))
```
