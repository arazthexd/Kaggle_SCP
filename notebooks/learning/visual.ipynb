### Matplotlib

#### Simple Plotting

```python
x = np.arange(1, 6)
y = x ** 1.5

fig = plt.figure(figsize=(3, 3)) # first is the x and second is the y
plt.plot(x, x, "r^", x, y, "bo")
plt.title("hihi")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
data = {
    "x1": np.arange(50),
    "x2": np.arange(50) ** 2 + np.random.randn(50) * 50,
    "y": np.random.randn(50),
    "d": np.random.rand(50) * 20
}

fig = plt.figure(figsize=(3, 3)) # first is the x and second is the y
plt.scatter("x1", "x2", c="y", s="d", data=data) # NOTE: s is size, c is color, x1, x2 are positional
plt.title("hihi")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### Figure

```python
# Creating and adding axes to figure:
fig = plt.figure(figsize=(2, 2))
fig.set_facecolor("darkgreen")
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1.1, 1.1, 1, 1])
ax1.scatter([1, 2, 3, 4], [3, 2, 2, 10])
ax2.plot([1, 2, 3, 4], [3, 2, 2, 10])
plt.show()

# Clear figure
# fig.clear()
```

```python
fig = plt.figure(facecolor="darkblue", figsize=(6,6)) 
axs = fig.subplots(2, 2)
fig.suptitle("oops", color="green")
axs[0, 0].set_xlim([-4, 4])
axs[0, 0].set_ylim([-3, 3])
plt.show()
```
